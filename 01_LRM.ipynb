```python
import matplotlib.pyplot as plt 
import numpy as np
from sklearn import datasets,linear_model
from sklearn.metrics import mean_squared_error
```

```python
diabetes = datasets.load_diabetes()
# print(diabetes.keys())
# diabetes_x = diabetes.data[:, np.newaxis,3]
diabetes_x = diabetes.data
# print(diabetes_x)
```

dict_keys(['data', 'target', 'frame', 'DESCR', 'feature_names', 'data_filename', 'target_filename', 'data_module'])

```python
diabetes_x_train = diabetes_x[29:]
diabetes_x_test = diabetes_x[:29]
```

```python
diabetes_y_train = diabetes.target[29:]
diabetes_y_test = diabetes.target[:29]
```

```python
model = linear_model.LinearRegression()
model.fit(diabetes_x_train,diabetes_y_train)
diabetes_y_predict = model.predict(diabetes_x_test)
```

```python
print(f"Mean squared error is{mean_squared_error(diabetes_y_test, diabetes_y_predict)}")
```

```python
print(f"WEIGHTS : {model.coef_}")
print(f"Intercepts : {model.intercept_}")
```

```python
plt.scatter(diabetes_x_test,diabetes_y_test)
plt.plot(diabetes_x_test,diabetes_y_predict)
plt.show()
```
