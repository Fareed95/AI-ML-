```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data.csv')
data
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
```

```python
data['outlook'] = label_encoder.fit_transform(data['outlook'])
data['humidity'] = label_encoder.fit_transform(data['humidity'])
data['windy'] = label_encoder.fit_transform(data['windy'])
data['play'] = label_encoder.fit_transform(data['play'])
data
```

```python
features_columns = ['outlook','humidity','windy']
X = data[features_columns]
y = data.play
```

```python
X 
```

```python
y
```

```python
from sklearn.model_selection import  train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_test
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
classifier = DecisionTreeClassifier(criterion ='gini')
```

```python
classifier.fit(X_train,y_train)
```

```python
classifier.predict(X_test)
```

```python
classifier.score(X_test,y_test)
```

```python
plot_tree(classifier)
```
