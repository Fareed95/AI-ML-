```python
import numpy as np 
import pandas as pd 
import  matplotlib.pyplot as plt 
```

```python
data = pd.read_csv('term_insurance_data.csv')
data
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
data['Bought Insurance'] = label_encoder.fit_transform(data['Bought Insurance'])
data
```

```python
plt.scatter(x='Age', y='Bought Insurance',data=data)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test,y_train,y_test = train_test_split(data[['Age']], data['Bought Insurance'])
```
