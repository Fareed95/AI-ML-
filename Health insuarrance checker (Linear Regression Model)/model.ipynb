```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
from sklearn import  linear_model
```

```python
data = pd.read_csv('health_insurance_random.csv')
data
```

```python
sns.lmplot(x='Age', y='Policy Cost', data=data)
```

```python
reg = linear_model.LinearRegression()
```

```python
reg_age =reg.fit(data[['Age']],data['Policy Cost'])
reg_policy_cost = reg.fit(data[['Policy Cost']], data['Age'])
```

```python
reg_age.predict([[23]])
```
